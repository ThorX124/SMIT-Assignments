# **PART 1: EASY LEVEL:-**

```python
# 1:
fruits:  list= ['Apple', 'Banana', 'Orange']
print(fruits)

# 2:
print(fruits[1])

# 3:
fruits[2] = 'Mango'
print(fruits)

# 4:
nums: tuple = (1, 2, 3, 4)
print(nums)

# 5:
print(len(nums))
```

# **PART 2: MEDIUM LEVEL:-**

```python
# 6:
fruits:  list= ['Apple', 'Banana', 'Orange']
fruits.append('Mango')
print(fruits)

# 7:
fruits.remove("Mango")
print(fruits)

# 8:
numbers: list = list(nums)
print(numbers)

# 9:
student: dict = {'name': 'Anas', 'age': 19, 'course': 'Agentic AI'}

# 10:
print(student['course'])
```

# **PART 3: HARD LEVEL:-**

```python
# 11:
for fruit in fruits:
  print(fruit)

# 12:
if 2 in numbers:
  print(True)

else:
  print(False)

# 13:
student['grade'] = 'A+'
print(student)

# 14:
student: dict = {'name': 'Anas', 'age': 19, 'course': 'Agentic AI'}
del student['age']
print(student)

# 15:
final: list = fruits + numbers
print(final)
```
