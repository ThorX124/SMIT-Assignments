# **TASK 1: CONDITIONAL STATEMENTS:-**

```python
# 1.1: Divisiblity Checker:-

num = int(input("Enter a number: "))    # Taking integer input by the user.

if num % 3 == 0:    # Checking if the number is divisible by 3.
  print(f'{num} is divisible by 3')
else:
  print(f'{num} is not divisible by 3')
```

```python
# 1.2: Even/Odd with sign detection:-

num= int(input("Enter a number: "))    # Taking integer input by the user.

if num > 0:   # This if block is checking if the number is positive.
  if num % 2 == 0:    # This if block is checking if the number is even/odd.
    print('Positive even number')
  else:
    print('Positive odd number')

elif num < 0:
  if num % 2 == 0:   # This if block is checking if the number is even/odd.
    print('Negative even number')
  else:
    print('Negative odd number')

else:   # If the nuber is neither greater than zero nor less than zero, it is zero.
  print('Number is Zero')
```

```python
# 1.3: Grade Evaluator:-

marks = int(input("Enter your marks: "))    # Taking integer input by the user.

while marks > 100 or marks < 0:    # Resisting an error to stop the program. If someone enters number greater than 100 or less than zero, the program alerts and asks for a new number instead of giving a grade.
  print('Invalid Input!')
  print('Marks cannot be greater than 100.')
  marks = int(input("Please re-enter your marks: "))

else:   # It runs when the loop condition is false.
  if marks >= 90:
    print('Grade: A')
    print('Excellent!')

  elif marks >= 75:
    print('Grade: B')
    print('Good!')

  elif marks >= 50:
    print('Grade: C')
    print('Needs Improvement')

  elif marks >= 35:
    print('Grade: D')
    print('Needs Improvement')

  else:
    print('Grade: F')
    print('Failed!')
```

```python
# 1.4: Leap year checker:

year = int(input("Enter a year: "))   # Asking for the year.

if year % 4 == 0:   # Cheaking for the first rule for declaring leap year.
  print(f'{year} is a leap year.')
elif year % 100 == 0 and year % 400 == 0:   # Cheaking for the second rule for declaring leap year.
  print(f'{year} is a leap year.')
else:   # If both rules fail, its not a leap year.
  print(f'{year} is not a leap year.')

```

# **TASK 3: LOOP CONTROL:-**

```python
# 3.1: Divisiblity search with K:

lis: list = [15, 18, 19, 2, 200, 69, 786, 28, 92]   # Gave a list myself because I don't know how to get a list as input.
k: int = int(input('Enter a number: '))   # Asking for the number we have to search for divisiblity.

for num in lis:
  if num % k == 0:
    print(f'{num} is divisible by {k}')
    break

  else:
    print(f'No number in the list is divisible by {k}')
    break
```

```python
# 3.2: Skip Negative Numbers using continue:

lis: list = [-15, 18, -19, -2, 200, -69, 786, -28, -92]   # Gave a list myself because I don't know how to get a list as input.

for num in lis:
  if num < 0: # Will skip all negative numbers.
    continue

  else:
    print(num)
```
